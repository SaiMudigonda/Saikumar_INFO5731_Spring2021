```python
# Please write your code here

import requests
from bs4 import BeautifulSoup
import pandas as pd
Main_data = [] 
Sub_data =[] 
for number in range(25):
  # Giving link directly
  link = "https://www.flipkart.com/search?q=samsung&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off" + str(number)
# Webpage Accessing
  web_page = requests.get(link) 
  soup = BeautifulSoup(web_page.text, 'html.parser')

  # To get Product name by using the class name
  product_name = soup.find_all(class_='_4rR01T') 

  # To get the Product rating by using the class name
  product_rating = soup.find_all(class_='_3LWZlK') 

  # Iterating through the list
  for ele, sub_ele in zip(product_name, product_rating) : 
      Main_data.append(ele.text) 
      Sub_data.append(sub_ele.text)
      
      # dataframe for products
df = pd.DataFrame(list(zip(Main_data, Sub_data)), columns =['Product Name', 'Rating'])  
df
```

```python
#Coverting product name to lowercase
df['After Covernting to lowercase'] = df['Product Name'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

```python
#Removing punctuations
df['After Covernting to lowercase'] = df['After Covernting to lowercase'].str.replace('[^\w\s]','')
df
```

```python
#Removing Special Characters
import re
df['After Covernting to lowercase'] = df['After Covernting to lowercase'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
df
```

```python
#Removing numbers
df['After Covernting to lowercase'] = df['After Covernting to lowercase'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

```python
import nltk
nltk.download()
```

```python
#Correcting Spelling

from textblob import TextBlob
df['After Covernting to lowercase'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
#Lemitization
from textblob import Word
import nltk
nltk.download('wordnet')

df['After Covernting to lowercase'] = df['After Covernting to lowercase'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```python
#POS Tag Features
from nltk.tokenize import word_tokenize
pos = []
for sentence in df['After Covernting to lowercase']:
  text = word_tokenize(sentence)
  pos.append(nltk.pos_tag(text))
pos
```

```python
Adverb = []
CordinatingConjunction = []
Interjection = []
Noun = []
Verb = []
PersonalPronoun = []
predeterminer = []
Adjective = []
SubordinatingConjuction = []
Determiner = []

for value in pos:
 
  AdverbCount = 0
  CordinatingConjunctionCount = 0
  SubordinatingConjuctionCount = 0
  InterjectionCount = 0
  NounCount = 0
  VerbCount = 0
  PersonalPronounCount = 0
  predeterminerCount = 0
  AdjectiveCount = 0
  DeterminerCount = 0
  for word,tag in value:
    if tag == 'JJ':
      AdjectiveCount = AdjectiveCount + 1
    elif tag == 'RB':
      AdverbCount = AdverbCount + 1
    elif tag == 'CC':
      CordinatingConjunctionCount = CordinatingConjunctionCount + 1
    elif tag == 'UH':
      InterjectionCount = InterjectionCount + 1
    elif tag == 'NN':
      NounCount = NounCount + 1
    elif tag == 'VR':
      VerbCount = VerbCount + 1
    elif tag == 'PRP':
      PersonalPronounCount = PersonalPronounCount + 1
    elif tag == 'PDT':
      predeterminerCount = predeterminerCount + 1
    elif tag == 'DT':
      DeterminerCount = DeterminerCount + 1
    elif tag == 'IN':
      SubordinatingConjuctionCount = SubordinatingConjuctionCount + 1
  
  Adverb.append(AdverbCount)
  CordinatingConjunction.append(CordinatingConjunctionCount)
  Interjection.append(InterjectionCount)
  Noun.append(NounCount)
  Verb.append(VerbCount)
  PersonalPronoun.append(PersonalPronounCount)
  predeterminer.append(predeterminerCount)
  Adjective.append(AdjectiveCount)
  Determiner.append(DeterminerCount)
  SubordinatingConjuction.append(SubordinatingConjuctionCount)

df['Number of Adverbs'] = Adverb
df['Number of Nouns'] = Noun
df['Number of Verbs'] = Verb
df['Number of Personal Pronouns'] = PersonalPronoun
df['Number of Predeterminers'] = predeterminer
df['Number of Adjectives'] = Adjective
df['Number of Determiners'] = Determiner
df['Number of Cordinating Conjunctions'] = CordinatingConjunction
df['Number of Subordinating Conjuctions'] = SubordinatingConjuction
df['Number of Interjections'] = Interjection
df
```
